```python
import numpy as np
from scipy.integrate import solve_bvp
from matplotlib import pyplot as plt
```

```python
#21.9

r, N0, delta, tf = 1, 2, .5, 5
pf = -1
def f(t, y, p):
    #N' = -rNlog(N) - duN, u = - pdeltaN/2
    #p' = pr(log(N) + 1) + d^2p^2N/2
    
    u = delta*y[0]*y[1]/2
    return p[0]*np.array([
        -r*y[0]*np.log(y[0]) + delta*u*y[0], 
        y[1]*r*(np.log(y[0]) + 1) + delta*delta*y[1]*y[1]*y[0]/2
    ])

def bc(ya, yb, p):
    ub = -yb[0]*yb[1]/(2*delta)
    Nb, pb = yb
    return np.array([
        ya[0] - N0, #N(0) = N0
        yb[1] - pf, #p(t_f) = pf
        pb*Nb*(r*np.log(Nb) + delta*ub) + ub**2 -.75* yb[0], #H(t_f) = .75 N(t_f)
    ])
t = np.linspace(0,1,100)
guess = np.ones(shape=(2, 100))
p_guess = np.ones(1)
sol = solve_bvp(f, bc, t, guess, p_guess)
print(sol.p)
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(sol.x, sol.y[0])
plt.xlabel("t")
plt.ylabel("N(t)")
plt.title("Optimal state (cancer cells)")
plt.subplot(1,2,2)
plt.plot(sol.x, -delta*sol.y[0]*sol.y[1]/2)
plt.xlabel("t")
plt.ylabel("u(t)")
plt.title("Optimal control (drug administered)")
plt.show()
```
